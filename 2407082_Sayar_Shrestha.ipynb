**Task - 3 - Outlier Detection:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

2. Define outliers using the 1.5  IQR rule.

```python
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

dataframe_copy = dataframe.copy()

south_asia_df = dataframe_copy[dataframe_copy['Country name'].isin(south_asian_countries)]


q1_happiness = south_asia_df['score'].quantile(0.25)
q3_happiness = south_asia_df['score'].quantile(0.75)
iqr_happiness = q3_happiness - q1_happiness

lower_bound_happiness = q1_happiness - 1.5 * iqr_happiness
upper_bound_happiness = q3_happiness + 1.5 * iqr_happiness

q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
iqr_gdp = q3_gdp - q1_gdp

lower_bound_gdp = q1_gdp - 1.5 * iqr_gdp
upper_bound_gdp = q3_gdp + 1.5 * iqr_gdp

outliers_df = south_asia_df[
    (south_asia_df['score'] < lower_bound_happiness) |
    (south_asia_df['score'] > upper_bound_happiness) |
    (south_asia_df['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_df['Log GDP per capita'] > upper_bound_gdp)
]

outliers_df
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
plt.figure(figsize=(8, 6))

plt.scatter(south_asia_df[~south_asia_df.index.isin(outliers_df.index)]['Log GDP per capita'],
            south_asia_df[~south_asia_df.index.isin(outliers_df.index)]['score'],
            color='blue', label='Non-Outliers')

plt.scatter(outliers_df['Log GDP per capita'],
            outliers_df['score'],
            color='red', label='Outliers')

plt.title('Scatter Plot of Score vs. GDP per capita (South Asia)')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

**Task - 4 - Exploring Trends Across Metrics:**

1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.

2. Create scatter plots with trendlines for these metrics against the Score.

3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

dataframe_copy = dataframe.copy()

south_asia_df = dataframe_copy[dataframe_copy['Country name'].isin(south_asian_countries)]

freedom_correlation = south_asia_df['score'].corr(south_asia_df['Freedom to make life choices'])
generosity_correlation = south_asia_df['score'].corr(south_asia_df['Generosity'])

print(f"Correlation between Freedom to Make Life Choices and Score: {freedom_correlation}")
print(f"Correlation between Generosity and Score: {generosity_correlation}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(x='Freedom to make life choices', y='score', data=south_asia_df, color='blue', line_kws={'color': 'red'})
plt.title('Freedom to Make Life Choices vs. Score')

plt.subplot(1, 2, 2)
sns.regplot(x='Generosity', y='score', data=south_asia_df, color='green', line_kws={'color': 'red'})
plt.title('Generosity vs. Score')

plt.tight_layout()
plt.show()

print("\nDiscussion:")
print("Based on the correlation coefficients and scatter plots, we can observe the following relationships:")

if abs(freedom_correlation) > abs(generosity_correlation):
    print("- Freedom to Make Life Choices has a stronger relationship with Score compared to Generosity.")
    print(f"  - The correlation coefficient for Freedom to Make Life Choices is {freedom_correlation}, indicating a {'positive' if freedom_correlation > 0 else 'negative'} relationship.")
    print(f"  - The correlation coefficient for Generosity is {generosity_correlation}, indicating a {'positive' if generosity_correlation > 0 else 'negative'} relationship.")
else:
    print("- Generosity has a stronger relationship with Score compared to Freedom to Make Life Choices.")
    print(f"  - The correlation coefficient for Generosity is {generosity_correlation}, indicating a {'positive' if generosity_correlation > 0 else 'negative'} relationship.")
    print(f"  - The correlation coefficient for Freedom to Make Life Choices is {freedom_correlation}, indicating a {'positive' if freedom_correlation > 0 else 'negative'} relationship.")

print("The scatter plots with trendlines provide a visual representation of these relationships. A steeper trendline indicates a stronger relationship.")
print("Further analysis can be done to explore the reasons behind these relationships and their implications for happiness in South Asian countries.")
```

**Task - 5 - Gap Analysis:**

1. Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score
for each South Asian country.

2. Rank the South Asian countries by this gap in both ascending and descending order.

3. Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.

4. Analyze the reasons behind these gaps and their implications for South Asian countries.

```python
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Create a copy of the DataFrame before filtering
dataframe_copy = dataframe.copy()

# Filter the copied dataframe to include only South Asian countries
south_asia_df = dataframe_copy[dataframe_copy['Country name'].isin(south_asian_countries)]

# --- 1. Calculate GDP-Score Gap ---
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

# --- 2. Rank by Gap ---
# Ascending order (smallest gaps first)
ranked_ascending = south_asia_df.sort_values(by=['GDP-Score Gap'])

ranked_descending = south_asia_df.sort_values(by=['GDP-Score Gap'], ascending=False)

# --- 3. Visualize Top Gaps ---
top_3_positive_gap = ranked_descending.head(3)
top_3_negative_gap = ranked_ascending.head(3)

plt.figure(figsize=(10, 6))
plt.bar(top_3_positive_gap['Country name'], top_3_positive_gap['GDP-Score Gap'], color='skyblue', label='Positive Gap')
plt.bar(top_3_negative_gap['Country name'], top_3_negative_gap['GDP-Score Gap'], color='salmon', label='Negative Gap')
plt.title('Top 3 Countries with Largest Positive and Negative GDP-Score Gaps')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()

# --- 4. Analysis and Discussion ---
print("\nAnalysis and Discussion:")
print("The GDP-Score Gap represents the difference between a country's economic performance (GDP per capita) and its overall happiness score.")
print("A positive gap indicates that a country's GDP per capita is higher than what might be expected based on its happiness score, suggesting potential underperformance in happiness relative to economic achievements.")
print("A negative gap indicates that a country's happiness score is higher than what might be expected based on its GDP per capita, suggesting potential overperformance in happiness relative to economic achievements.")

print("\nTop 3 countries with the largest positive gaps:")
for index, row in top_3_positive_gap.iterrows():
    print(f"- {row['Country name']}: GDP-Score Gap = {row['GDP-Score Gap']:.2f}")

print("\nTop 3 countries with the largest negative gaps:")
for index, row in top_3_negative_gap.iterrows():
    print(f"- {row['Country name']}: GDP-Score Gap = {row['GDP-Score Gap']:.2f}")

print("\nPossible reasons for these gaps could include:")
print("- Social factors: Strong social support systems, cultural values, and community bonds can contribute to higher happiness even with lower GDP per capita.")
print("- Governance and institutions: Effective governance, low corruption, and access to quality public services can enhance happiness.")
print("- Environmental factors: Clean environment, access to nature, and sustainable practices can positively impact well-being.")
print("- Inequality and distribution of wealth: High levels of inequality can negatively affect happiness even in countries with high GDP per capita.")

print("\nImplications for South Asian countries:")
print("- Countries with large positive gaps should focus on improving social and institutional factors to enhance happiness alongside economic growth.")
print("- Countries with large negative gaps can leverage their strengths in social and institutional factors to maintain high happiness levels while pursuing economic development.")
print("- Understanding the factors contributing to these gaps can help policymakers develop targeted strategies to improve overall well-being in the region.")
```
